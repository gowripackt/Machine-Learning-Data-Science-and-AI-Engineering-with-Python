# Prerequisites

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
```

# Prepare the data

```python
KAGGLE = True
if KAGGLE:
    data_path = "/kaggle/input/titanic"
else:
    data_path = "data"
        
train_df = pd.read_csv(f"{data_path}/train.csv")
test_df = pd.read_csv(f"{data_path}//test.csv")
```

# Feature engineering

```python
for dataset in [train_df, test_df]:
    dataset['Sex'] = dataset['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
```

# Prepare the training

```python
VALID_SIZE = 0.2
train, valid = train_test_split(train_df, test_size=VALID_SIZE, random_state=42, shuffle=True)
```

```python
predictors = ["Sex", "Pclass"]
target = 'Survived'
```

```python
train_X = train[predictors]
train_Y = train[target].values
valid_X = valid[predictors]
valid_Y = valid[target].values
```

# Training

```python
clf = RandomForestClassifier(n_jobs=-1, 
                             random_state=42,
                             criterion="gini",
                             n_estimators=100,
                             verbose=False)
```

```python
clf.fit(train_X, train_Y)
```

# Evaluation

Let's first evaluate the model by predicting using training data.

```python
preds_tr = clf.predict(train_X)
```

```python
print(metrics.classification_report(train_Y, preds_tr, target_names=['Not Survived', 'Survived']))
```

Using now the evaluation set:

```python
preds = clf.predict(valid_X)
```

```python
print(metrics.classification_report(valid_Y, preds, target_names=['Not Survived', 'Survived']))
```
